```python
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
import datetime as dt
import matplotlib as mpl
import pandas as pd
plt.style.use('classic')
```

### Chaging Matplotlib Default Paramenters

[Matplotlib Reference](https://matplotlib.org/users/customizing.html)

```python
# Let us save a copy of the default parameters just in case

Ipython_default = plt.rcParams.copy()
```

```python
from matplotlib import cycler

colors = cycler('color',
               ['#EE6666', '#3388BB', '#9988DD',
                 '#EECC55', '#88BB44', '#FFBBBB'])

plt.rc('axes', facecolor='E6E6E6',edgecolor='none',axisbelow=True,
      grid=True,prop_cycle=colors)
plt.rc('grid',color='w',linestyle='solid')
plt.rc('xtick',direction='out',color='gray')
plt.rc('ytick',direction='out',color='gray')
plt.rc('patch',edgecolor='E6E6E6')
plt.rc('lines',linewidth=2)

x = np.random.randn(1000)
plt.hist(x);
```

```python
for i in range(4):
    plt.plot(np.random.rand(10))
```

### Matplotlib Style Sheet
------------------------------

Play around with the different styles available 

```python
plt.style.available
```

```python
def hist_and_lines():
    np.random.seed(0)
    fig, ax = plt.subplots(1, 2, figsize=(11, 4))
    ax[0].hist(np.random.randn(1000))
    for i in range(3):
        ax[1].plot(np.random.rand(10))
    ax[1].legend(['a', 'b', 'c'], loc='lower left')
```

### Applying a different style only to an specific chart

```python
# reset rcParams
plt.rcParams.update(Ipython_default);

hist_and_lines()
```

```python
with plt.style.context('seaborn-deep'):
    hist_and_lines()
```

```python
hist_and_lines()
```

```python
with plt.style.context('bmh'):
    hist_and_lines()
```

```python
with plt.style.context('dark_background'):
    hist_and_lines()
```
